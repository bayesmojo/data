```python
import pandas as pd

data = pd.read_csv("HC.csv")
```

```python
data = data.to_numpy()
A = data[:, 0]
B = data[:, 1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(A, B)
# plt.savefig("hct.pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(A, B)
tex = range(1, 11) 
for label, x, y in zip(tex, A, B):
    plt.annotate( 
        label,
        xy=(x, y), 
        xytext=(-6, -4), 
        textcoords='offset points', ha='right', va='bottom') 
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
from yellowbrick.cluster import KElbowVisualizer
mod = AgglomerativeClustering()

visualizer = KElbowVisualizer(mod, k=(1, 10), timings=False) 
visualizer.fit(data)
visualizer.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
label = model.fit_predict(data)
print("clustering=", label)
```

On the above elbow diagram, the distortion score is the sum of the squared distances between each observation vector and its dominating centroid.

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

label = label
colors = ['red','green']
plt.scatter(A, B, c=label, cmap=ListedColormap(colors))
plt.axis("scaled") 


tex = range(1, 11)
for label, x, y in zip(tex, A, B):
    plt.annotate(
        label,
        xy=(x, y), 
        xytext=(-6, -4),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```

```python
import pandas as pd
import numpy as np 
from sklearn.cluster import AgglomerativeClustering

# precomputed distance matrix
X = np.array([[0, 0.85, 0.65, 0.13], 
              [0.85, 0, 0.83, 0.71], 
              [0.65, 0.83, 0, 0.79], 
              [0.13, 0.71, 0.79, 0]])
print("distance matrix =", '\n' , X)

model = AgglomerativeClustering(n_clusters=2, affinity='precomputed', linkage='single')  

label = model.fit_predict(X)
print("clustering=", '\n', label)
```

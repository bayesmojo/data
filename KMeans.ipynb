```python
import pandas as pd

data = pd.read_csv("HC.csv")
```

```python
data = data.to_numpy()
A = data[:, 0]
B = data[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(A, B)
plt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
mod = KMeans()

visualizer = KElbowVisualizer(mod, k=(1, 10), timings=False) 
visualizer.show()
```

```python
from sklearn.cluster import KMeans # K is capital K
model = KMeans(n_clusters=2).fit(data)
```

```python
label = model.predict(data)
print("clustering=", label)
```

```python
model.cluster_centers_ 
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

label = label
colors = ['red','green']
plt.scatter(A, B, c=label, cmap=ListedColormap(colors))
plt.axis("scaled") 

tex = range(1, 11)    
for label, x, y in zip(tex, A, B):       
    plt.annotate(
        label,
        xy=(x, y), 
        xytext=(-6, -4),
        textcoords='offset points', ha='right', va='bottom')
plt.show()
```
